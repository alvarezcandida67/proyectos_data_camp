```python
# Analice los datos de los ganadores del Premio Nobel e identifique los patrones respondiendo a las siguientes preguntas:

# Librerías necesarias
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Cargamos el dataset
nobel = pd.read_csv('nobel.csv')
nobel.head()
```

```python
# 1. ¿Cuál es el género mas frecuente y el país de nacimiento más comunmente galardonado?

top_gender = nobel['sex'].value_counts().index[0]
top_country = nobel['birth_country'].value_counts().index[0]

print('93% de los galardonados fueron hombres. EE. UU. es el país con más Premios Nobel con 30% de ellos.')
```

```python
# 2. ¿Qué década tuvo la mayor proporción de ganadores del Premio Nobel nacidos en EE. UU. y  en todas las categorías?

# Para establecer la columna de decadas, debemos sustituir el ultimo digito del año por un 0. Por ejemplo 1986 -> 1980
nobel['decade'] = nobel['year'].astype(str).str[:-1] + '0'
nobel['decade'] = nobel['decade'].astype(int)

# Ahora debemos filtrar los datos para los ganadores nacidos en EE. UU. y luego agrupar por década
usa_winners = nobel[nobel['birth_country'] == 'United States of America']
grouped_data = usa_winners.groupby('decade', as_index=False).count()

# Ubicamos el indice del valor mas alto de la columna 'prize'
indice_max_prize = grouped_data['prize'].idxmax()

# Aplicamos ese indice al valor de la columna 'decade'
max_decade_usa = grouped_data.loc[indice_max_prize, 'decade']

print('La decada del',max_decade_usa,'tuvo el mayor numero de Premios Nobel nacidos en EE. UU.')
```

```python
# 3. ¿Qué combinación de la década y la categoría del Premio Nobel tuvo la mayor proporción de mujeres laureadas?

# Agregamos una columna de booleanos para identificar los Nobel ganados por mujeres
nobel['fem_winner'] = nobel['sex'] == 'Female'

# Agrupamos por década y categoría y calculamos el número total de premios ganados por mujeres y el número total de premios en cada categoría
female_winners_3 = nobel[nobel['fem_winner']].groupby(['decade', 'category']).size()
total_winners = nobel.groupby(['decade', 'category']).size()

# Calculamos la proporción de mujeres laureadas en cada década y categoría
fem_prop = female_winners_3 / total_winners

# Obtenemos la combinación con la mayor proporción de mujeres laureadas
max_fem_prop_combination = fem_prop.idxmax()

print("El año y categoría con la mayor proporción de mujeres laureadas es:", max_fem_prop_combination)
```

```python
# 4. ¿Quién fué la primera mujer en ganar un Premio Nobel?

# Necesitamos aislar las filas de 'nobel' donde ganaron mujeres.
female_winners_4 = nobel[nobel['fem_winner']]

# Ordenamos los registros de mujeres ganadoras por año, de forma ascendente. 
female_winners_sorted = female_winners_4.sort_values(by='year')

# Seleccionamos el primer registro de la lista resultante
first_woman_name = female_winners_sorted.iloc[0]

print("La primera mujer en ganar un Premio Nobel fue:", first_woman_name['full_name'], 'en la categoría:', first_woman_name['category'])
```

```python
# 5. ¿Qué personas u organizaciones han ganado más de un Premio Nobel a lo largo de los años?

# Agrupamos los datos por nombre completo y contamos el número de premios para cada persona u organización
nobel_winners_count = nobel.groupby('full_name').size()

# Aquellos que han ganado más de un premio:
nobel_multiple_winners = nobel_winners_count[nobel_winners_count > 1]

# Ordenamos los resultados de mayor a menor
nobel_multiple_winners_sorted = nobel_multiple_winners.sort_values(ascending=False)


repeat_list = list(nobel_multiple_winners_sorted.index)
```
